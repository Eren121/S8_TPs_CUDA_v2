# 3) Profilage

Pour vérifier si les programmes fonctionnent bien, on peut vérifier l'intégrité de la mémoire grâce à __cuda-memcheck__.

```python
%%bash
cd ..

cuda-memcheck ./add2vec.out
cuda-memcheck ./farhenheit.out 1000
cuda-memcheck ./bw.out 100
cuda-memcheck ./blur.out 100
cuda-memcheck ./matmul.out 100
```

Maintenant on peut mesurer les performance effectives sur GPU grâce à __nvprof__.

```python
%%bash
cd ..


nvprof --print-gpu-trace ./matmul.out 300 --benchmark -numdevices=2 -i=1 2>&1 > /dev/null
echo -e "\n\n#################################\n\n"
nvprof --print-gpu-trace ./blur.out 300 --benchmark -numdevices=2 -i=1 2>&1 > /dev/null
```
